## **I was interested in knowing how often each Age_Groups purchased products. So i decided to do the count of every Age_Groups**

```python
y= bs.groupby('Age_Group').agg({'Age_Group':'count'}).rename(columns = {'Age_Group':'Group','Age_Group':'Count'})


y = y.sort_values(by = 'Count', ascending= False)

x = y.index
y = y['Count']
y
```

```python
bar = plt.bar(x,y)
plt.title(label='Age_Group Count', fontsize='18')
plt.xlabel(xlabel='Age Group', fontsize='16')
plt.ylabel(ylabel='Count', fontsize='16')
plt.bar_label(bar,label= y ,label_type='edge')
plt.show()
```

```python
bs.head(2)
```

## **One of the questions chatgpt created was for me to provide the Average revenue by year. So that was what i decided to plot in the next few cells**

```python
y = bs.groupby('Year').agg({'Revenue':'mean'}).rename(columns = {'Year':'Year','Revenue':'Revenue'})
y = round(y)

y = y.reset_index().sort_values(by='Revenue', ascending=True)

x = y['Year']
y = y['Revenue'].values

```

```python
year_color = {'2017':'darkgreen','2018':'mediumseagreen','2019':'mediumspringgreen','2020':'seagreen','2021':'green'}
colors = [year_color[year] for year in x]
plt.figure(figsize=(10,6))
bar = plt.bar(x,y, color = colors)
plt.title('Average Revenue by Year', fontsize = '20')
plt.xlabel('Year', fontsize = '16')
plt.bar_label(bar,label= y ,label_type='edge',color = 'k', fontweight = 'bold', padding = 2)
plt.show()
```

# **## I wanted to know what product_category made the biggest profit. After that, i was curious to check how the bike sub_categories contributes to it being the most profitable product_category**

```python
y= bs.groupby('Product_Category').agg({'Profit':sum}).rename(columns = {'Product_Category':'Product_Category','Profit':'Profit'})
y = y.reset_index()

x = y['Product_Category']
y = y['Profit']
```

```python
plt.pie(y, labels=x, autopct='%1.1f%%')
plt.title('Profit by Product Category', fontsize = '20')
plt.show()
```

```python
bs.head(2)
```

```python
bs2 = bs[bs.Product_Category == 'Bikes']
bs2.Sub_Category.unique()
```

```python
y = bs2.groupby('Sub_Category').agg({'Profit':sum}).rename(columns = {'Sub_Category':'Sub_Category','Profit':'Profit'})
y = y.reset_index()

x = y['Sub_Category']
y = y['Profit']
```

```python
plt.pie(y, labels=x, autopct='%1.1f%%', startangle = 90)
plt.title('Profit by Sub Category', fontsize = '20')
plt.show()
```

# **After that, I decided to check the overall sub_category and how much revenue they bring as a group...**

```python
bs.Sub_Category.nunique()
```

```python
y = bs.groupby('Sub_Category').agg({'Revenue': 'mean'}).rename(columns = {'Sub_Category':'Sub_Category','Revenue':'Revenue'})
y = y.reset_index().sort_values(by='Revenue', ascending=True)
y = round(y)

x = y['Sub_Category']
y = y['Revenue'].values

y
```

```python
fig, ax = plt.subplots(figsize = (10,6))

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.xaxis.set_visible(False)

barh = plt.barh(x,y, height = 0.5, color = 'yellowgreen')
plt.title('Revenue by Sub Category', fontsize = '20')
plt.bar_label(barh,label= y ,label_type='edge',color = 'k', fontweight = 'bold', padding = 2)
plt.show()
```
